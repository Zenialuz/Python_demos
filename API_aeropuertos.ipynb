```python
# pero para tener la lista de todos los aeropuestos de todos los paises hacemos lo siguiente:

url_paises='https://www.air-port-codes.com/api/v1/countries'

resultado_paises = req.post(url_paises,headers=cabeceras)

datos_paises= resultado_paises.json()
datos_paises

```

```python
paises=datos_paises['countries']

paises
```

```python
total_aeropuertos=[]

for pais in paises:
    
    parametros = {
    'term':pais['name']
    }
    resultado=req.post(url,headers=cabeceras,params=parametros)
    
    if resultado.status_code==200:
    
        datos = resultado.json()
        
        if datos['statusCode']==200:
            aeropuertos=datos['airports']
            total_aeropuertos.extend(aeropuertos)
total_aeropuertos
    
```

```python
total_aeropuertos_normalizado = pd.json_normalize(total_aeropuertos)

total_aeropuertos_normalizado
```

```python
total_aeropuertos_normalizado=total_aeropuertos_normalizado.applymap(str)
```
